## <font color="maroon"><h4 align="center">Handling Missing Data - replace method</font>

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data.csv")
df
```

**Replacing single value**

```python
new_df = df.replace(-99999, value=np.nan)
new_df
```

**Replacing list with single value**

```python
new_df = df.replace(to_replace=[-99999,-88888,'0'], value=np.nan)
new_df
```

**Replacing per column**

```python
df
```

```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)
new_df
```

**Replacing by using mapping**

```python
new_df = df.replace({
        -99999: np.nan,
        '0': 'No Event',
    })
new_df
```

**Replacing list with another list**

```python
df = pd.DataFrame({
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica'],
    'grade': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'gender':['m', 'f', 'm','m','f','f']
})
df
```

```python
df['grade'].unique()
```

```python
df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4], inplace = True)
```
